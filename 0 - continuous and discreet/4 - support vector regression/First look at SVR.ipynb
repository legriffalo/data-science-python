```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Support vectorisation requires feature scaling
from sklearn.preprocessing import StandardScaler

# allow us to use support vector regression
from sklearn.svm import SVR

from sklearn.model_selection import train_test_split
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

y = y.reshape(len(y),1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)


# scale the features
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
#svr has hyper parameters c and epsilon
# c is how much variance the fit line exhibits (reduces slack)
# epsilon is how far from the tube a data point can be before 
# it is used as a support vector
regression = SVR(kernel = 'rbf', C= 30, epsilon = 0.002)
regression.fit(X_train,y_train)
```

```python
sc_y.inverse_transform(regression.predict([[4]]))
```

```python
X = sc_X.fit_transform(np.arange(1,10,1).reshape(-1,1))

# Visualising the SVR results
plt.scatter(sc_X.inverse_transform(X_train), sc_y.inverse_transform(y_train), color = 'red')
plt.plot(sc_X.inverse_transform(X),sc_y.inverse_transform(regression.predict(X)), color = 'blue')        
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the SVR results (for higher resolution and smoother curve)
X_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))


plt.scatter(sc_X.inverse_transform(X_train), sc_y.inverse_transform(y_train), color = 'red')
plt.plot(X_grid, sc_y.inverse_transform(regression.predict(sc_X.transform(X_grid))), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
